## VISUALISATIONS

```python
df1 = df.iloc[:, 2:].reset_index(drop=True)
```

```python
df1
```

# Line Graphs

```python
# Group by 'Generation' and calculate the mean for each stat
generation_stats = df1.groupby('GENERATION').mean()[['HP', 'ATTACK', 'DEFENSE', 'SP. ATK', 'SP. DEF', 'SPEED']]
```

```python
generation_stats
```

```python
import matplotlib.pyplot as plt

# Plotting with Matplotlib
plt.figure(figsize=(12, 6))

# Plot each stat
for column in generation_stats.columns:
    plt.plot(generation_stats.index, generation_stats[column], marker='o', label=column)

# Adding title and labels
plt.title('Average Pokémon Stats by Generation')
plt.xlabel('Generation')
plt.ylabel('Average Stat Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Reset the index of generation_stats for seaborn plotting
generation_stats_reset = generation_stats.reset_index().melt(id_vars='GENERATION', var_name='Stat', value_name='AVERAGE VALUE')

# Plotting with Seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=generation_stats_reset, x='GENERATION', y='AVERAGE VALUE', hue='Stat', marker='o')

# Adding title and labels
plt.title('Average Pokémon Stats by Generation (Seaborn)')
plt.xlabel('Generation')
plt.ylabel('Average Stat Value')
plt.grid(True)
plt.show()
```

# Bar Graphs

# >   A bar chart to show the total count of Pokémon in each generation

```python
# Count the number of Pokémon in each generation
generation_counts = df['GENERATION'].value_counts().sort_index()

# Display the count of Pokémon per generation
generation_counts
```

```python
# Plotting with Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(generation_counts.index, generation_counts.values, color='skyblue')

# Adding title and labels
plt.title('Number of Pokémon by Generation')
plt.xlabel('Generation')
plt.ylabel('Number of Pokémon')
plt.xticks(generation_counts.index)
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Create a DataFrame for the generation counts
generation_counts_df = generation_counts.reset_index()
generation_counts_df.columns = ['Generation', 'Count']

# Plotting with Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=generation_counts_df, x='Generation', y='Count', palette='viridis')

# Adding title and labels
plt.title('Number of Pokémon by Generation (Seaborn)')
plt.xlabel('Generation')
plt.ylabel('Number of Pokémon')
plt.grid(True)
plt.show()
```

# Histogram

# Scatter

### Fire Vs Water

```python
# Filtering Fire and Water Pokémon
fire = df[(df['TYPE 1'] == 'Fire') | (df['TYPE 2'] == 'Fire')]
water = df[(df['TYPE 1'] == 'Water') | (df['TYPE 2'] == 'Water')]

# Creating the scatter plot
plt.scatter(fire[fire['LEGENDARY'] == False]['ATTACK'].head(50), fire[fire['LEGENDARY'] == False]['DEFENSE'].head(50), 
            c='r', label='Fire (Non-Legendary)', marker="*", s=50)
plt.scatter(fire[fire['LEGENDARY'] == True]['ATTACK'].head(50), fire[fire['LEGENDARY'] == True]['DEFENSE'].head(50), 
            c='darkred', label='Fire (Legendary)', marker="*", s=75)

plt.scatter(water[water['LEGENDARY'] == False]['ATTACK'].head(50), water[water['LEGENDARY'] == False]['DEFENSE'].head(50), 
            c='b', label='Water (Non-Legendary)', s=25)
plt.scatter(water[water['LEGENDARY'] == True]['ATTACK'].head(50), water[water['LEGENDARY'] == True]['DEFENSE'].head(50), 
            c='navy', label='Water (Legendary)', s=50)

# Adding labels and legend
plt.xlabel("Attack")
plt.ylabel("Defense")
plt.legend()

# Setting the size of the figure
fig = plt.gcf()
fig.set_size_inches(12, 6)

# Display the plot
plt.title("Scatter Plot of Attack vs. Defense for Fire and Water Pokémon")
plt.show()
```

This shows that fire type pokemons have a better attack than water type pokemons but have a lower defence than water type.

# Seaborn

```python
# Creating a combined DataFrame
combined_data = pd.concat([fire, water])

# Plotting with Seaborn
plt.figure(figsize=(12, 6))
sns.scatterplot(data=combined_data.head(100), x='ATTACK', y='DEFENSE', hue='Type', style='LEGENDARY', s=100)

# Adding labels and legend
plt.xlabel("Attack")
plt.ylabel("Defense")
plt.legend(title='Type and Legendary Status')
plt.title('Scatter Plot of Fire and Water Pokémon (Seaborn)')

# Display the plot
plt.show()
```

# Analysis and Interpretation:

**Distribution of Pokemon by Primary Type**:
   - This bar chart shows the number of Pokemon for each primary type.
   - **Insight**: The most common primary types are Water, Normal, and Grass, while the least common types are Flying, Ghost, and Ice.

**Bar Chart of total number of Pokemon in different generations**:
   - This bar chart shows the number of Pokemon for each primary type.
   - **Insight**: The most common primary types are Water, Normal, and Grass, while the least common types are Flying, Ghost, and Ice.


**Distribution of Pokemon by Total Stats:**

* What it Represents: This histogram illustrates the frequency distribution of Pokemon based on their total stats.
* **Insights**: The majority of Pokemon have total stats between 300 and 500. There are fewer Pokemon with very high or very low total stats, indicating a clustering around the mid-range stats. This suggests that most Pokemon are designed to be relatively balanced in terms of overall strength.


**Scatterplot of Attack vs Defense:**

* What it Represents: This scatterplot shows the relationship between Attack and Defense stats, distinguishing between Legendary and non-Legendary Pokemon.
* **Insights:** Legendary Pokemon tend to have higher Attack and Defense stats, as expected. There is no strong correlation between Attack and Defense overall, indicating diverse roles and strategies among Pokemon. This implies that some Pokemon are designed to be strong attackers with low defense, while others might be more balanced or defense-oriented.

**Heatmap of Correlation Matrix:**

* What it Represents: This heatmap shows the correlation between different Pokemon stats.
* **Insights:** There are strong positive correlations between Total and other individual stats (HP, Attack, Defense, Sp. Atk, Sp. Def, Speed). Notably, Sp. Atk and Sp. Def have a moderate positive correlation, suggesting that Pokemon with high special attack also tend to have high special defense. This might indicate that special-oriented Pokemon are designed to excel in both offense and defense in the special category.
